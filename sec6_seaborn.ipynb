```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.distplot(tips['total_bill'], bins=50)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.set(style="darkgrid")
sns.boxplot('day', 'total_bill', data=tips, hue='smoker')
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.violinplot('day', 'total_bill', data=tips, hue='smoker', split=True)
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.swarmplot(x='day', y='total_bill', data=tips, hue='sex')
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='PiYG')
```

```python
# matrix plots
flights = sns.load_dataset('flights')
flights.head()
```

```python
fp = flights.pivot_table(index='month', columns='year', values='passengers')
```

```python
sns.heatmap(fp, cmap='coolwarm')
```

```python
sns.clustermap(fp, cmap='coolwarm', standard_scale=1)
```

```python
# GRIDS
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.distplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.scatterplot, 'total_bill', 'tip')
```

```python
# Regression plots
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], 
            scatter_kws={'s':50})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips,  col='sex', row='time')
```

```python
# style and color
plt.figure(figsize=(6,5))
sns.set_style('ticks')
sns.set_context('poster')
sns.countplot(x='sex', data=tips)
```
